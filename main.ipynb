```python
df['Has Table booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})
df['Has Online delivery'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})

# Determine the percentage of restaurants that offer table booking and online delivery
table_booking_percentage = (df['Has Table booking'].sum() / len(df)) * 100
online_delivery_percentage = (df['Has Online delivery'].sum() / len(df)) * 100

print("Percentage of Restaurants Offering Table Booking: {:.2f}%".format(table_booking_percentage))
print("Percentage of Restaurants Offering Online Delivery: {:.2f}%".format(online_delivery_percentage))
```

```python
average_rating_with_booking = df[df['Has Table booking'] == 1]['Aggregate rating'].mean()
average_rating_without_booking = df[df['Has Table booking'] == 0]['Aggregate rating'].mean()

print("Average Rating of Restaurants with Table Booking: {:.2f}".format(average_rating_with_booking))
print("Average Rating of Restaurants without Table Booking: {:.2f}".format(average_rating_without_booking))

# Analyze the availability of online delivery among restaurants with different price ranges
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price range', y='Has Online delivery', data=df)
plt.xlabel('Price Range')
plt.ylabel('Has Online Delivery')
plt.title('Availability of Online Delivery among Different Price Ranges')
plt.show()
```

```python
length_of_ResName = []
length_of_address = []

def lenfind(x):
    for i in x:
        leng = len(i)
        length_of_ResName.append(leng)

def lenfind2(x):
    for i in x:
        leng = len(i)
        length_of_address.append(leng)
        
lenfind(df['Restaurant Name'])
lenfind2(df['Address'])
```

```python
df['Restaurent name Length'] = length_of_ResName
df['Restaurent address length'] = length_of_address
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df.drop(['Restaurant Name', 'Address', 'Locality', 'Rating text'], axis=1, inplace= True)
df.sample()
```

```python
df['Rating color'].value_counts()
```

```python
df['Rating color'] = df['Rating color'].replace({
    'White': 0,
    'Dark Green': 5,
    'Green': 4,
    'Yellow': 3,
    'Orange': 2,
    'Red': 1
})
df.sample()
```

```python
df['Currency'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df['Currency'] = encoder.fit_transform(df['Currency'])
df.sample()
```

```python
df['Cuisines'] = encoder.fit_transform(df['Cuisines'])
df.sample()
```

```python
df['City'] = encoder.fit_transform(df['City'])
df.sample()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
lr = LinearRegression()
```

```python
X = df.drop('Aggregate rating', axis= 1)
y = df['Aggregate rating']
```

```python
X = scaler.fit_transform(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=37, shuffle=True,)
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, y_pred)
print(f'Accuracy is: {100 * r_squared.round(2)}%')
```
